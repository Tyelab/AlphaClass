```python
#################### Setting up the file ################
# videoRoot = "your_input_folder"
#videoFile = root + "your_video.mp4"


# root should be the location of your videos 
root = 'C:\\Users\\lkeyes\\Documents\\AlphaClassForLaurel\\FOMOvids\\'
arr = os.listdir(root)

# outputImageFolder is where you will put your extracted frames
# this folder should contain two subfolders 'images' and 'labels'
# individual images will be extracted by this notebook and placed in the folder images
# labels will contain a single csv of the label for each each image.  (typically, this is the output from using makesense.ai website for labeling.)
# The expected form of the csv is as follows:
# column 1 is the label.  There should be no spaces between words.  Use CamelCase styling or underscores to separate words, if needed.
# columns 2 and 3 are the x,y-coordinates of the label point on the image
# column 4 is the image name associated with that label
# column 5 and 6 are the width and height of the image (in pixels)

# we want to put extracted images into the 'images' folder, so I include it in the file path like so:
outputImageFolder = 'C:\\Users\\lkeyes\\Documents\\AlphaClassForLaurel\\TrainingData-01-08-2025\\images\\'
os.makedirs(outputImageFolder)

# print out the videos in your folder
#print (arr)
# print out number of videos in folder
#print (len(arr))
```

```python
for i in range(1,len(arr)):
        
        #thisvid = "Cage46M3-OM.mp4"
        thisvid = arr[i]
        videoFile = root + thisvid 
        #video_name = videoFile.split("Cage", 1)[1]
        print(videoFile)
        
        video_name =thisvid.partition('.mp4')[0]
        #print(video_name)

        #outputImage = root + "your_output_folder/"
        vidcap = cv2.VideoCapture(videoFile)
        success,image = vidcap.read()
        print(success)

        #################### Setting up parameters ################

        seconds = 1 # selecting 1 frame per 60 seconds Nancy changed this from 60 to 1.5
        fps = vidcap.get(cv2.CAP_PROP_FPS) # Gets the frames per second
        fps=round(fps)
        multiplier = fps * seconds
        print (fps)
        print (multiplier)
        
        #################### Initiate Process ################
        counter=1000
        while success and counter<2000:
            frameId = int(round(vidcap.get(1))) #current frame number, rounded b/c sometimes you get frame intervals which aren't integers...this adds a little imprecision but is likely good enough
            success, image = vidcap.read()
            print(success)
            if frameId % multiplier == 0:

                cv2.imwrite(outputImageFolder + video_name + "_%d.jpg" % frameId, image)
                counter+=1
        vidcap.release()
        temp_string = "Done with " + video_name
        print (temp_string)
        print (counter)
```
